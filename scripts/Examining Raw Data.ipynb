## Examinging Data from RACMOv2.3p2

```python
racmo_erai_temp_path = f'{raw_data_directory}/RACMO/RACMO_ERAINT/t2m.KNMI-2001.ANT27.ERAINx_RACMO2.4.1.3H.nc'
racmo_erai_snow_path = f'{raw_data_directory}/RACMO/RACMO_ERAINT/snowfall.KNMI-2001.ANT27.ERAINx_RACMO2.4.1.3H.nc'
racmo_erai_melt_path = f'{raw_data_directory}/RACMO/RACMO_ERAINT/snowmelt.KNMI-2001.ANT27.ERAINx_RACMO2.4.1.3H.nc'

for path in [racmo_erai_temp_path,racmo_erai_snow_path,racmo_erai_melt_path]:
    cubelist = iris.load(path)
    for cube in cubelist:
        if cube.long_name == '2-m Temperature':
            racmo_erai_temp_cube = cube
        elif cube.long_name == 'Solid Precipitative Flux':
            racmo_erai_snow_cube = cube
        elif cube.long_name == 'Snow Melt Flux':
            racmo_erai_melt_cube = cube
```
